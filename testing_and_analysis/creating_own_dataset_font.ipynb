```python
import pandas as pd
from os import walk
```

```python
my_pathes = [[r'data\data_self_creation\font_color\black_baclground_font_white/', 'white', 'black_baclground'],
            [r'data\data_self_creation\font_color\mixed_background_font_white/', 'white', 'mixed_background'],
            [r'data\data_self_creation\font_color\white_background_font_black/', 'black', 'white_background']]
```

```python

data_frame = []

for _tuple in my_pathes:
    
    path = _tuple[0]
    print(path)
    font_color = _tuple[1]
    dataset = _tuple[2]
    
    filenames = next(walk(path), (None, None, []))[2]  # [] if no file
    
    for file in filenames:
        
        data_frame.append({'path': path, 'file_name': file, 'font_color': font_color, 'dataset': dataset})
    
result_df = pd.DataFrame(data_frame)
result_df
```

```python
result_df.to_csv('own_dataset_font_color.csv')
```
